EJERICIO: EDADES- PESO

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------- GRFICO 1: MEDIA Y DESVIACIN ESTNDAR DE PESOS ----------
pesos = []
print("Ingrese el peso (en kg) de 5 estudiantes:")
for i in range(5):
    peso = float(input(f"Peso del estudiante #{i+1}: "))
    pesos.append(peso)

media = np.mean(pesos)
desviacion = np.std(pesos)

print(f"\nMedia de pesos: {media:.2f} kg")
print(f"Desviacin estndar de pesos: {desviacion:.2f} kg")

plt.figure(1)
valores_peso = [media, desviacion]
etiquetas_peso = [f'Media: {media:.2f}', f'Desviacin: {desviacion:.2f}']
colores_peso = ['lightblue', 'purple']

plt.pie(valores_peso, labels=etiquetas_peso, colors=colores_peso)
plt.title("Media y Desviacin Estndar del Peso")

# ---------- GRFICO 2: GNERO DE ESTUDIANTES ----------
generos = []
print("\nIngrese el gnero de estudiantes (M = masculino, F = femenino):")
for i in range(5):
    while True:
        genero = input(f"Gnero del estudiante #{i+1} (M/F): ").strip().upper()
        if genero in ['M', 'F']:
            generos.append(genero)
            break
        else:
            print("Entrada invlida. Use 'M' para varn o 'F' para mujer.")


varones = generos.count("M")
mujeres = generos.count("F")

print(f"\nCantidad de varones: {varones}")
print(f"Cantidad de mujeres: {mujeres}")

plt.figure(2)
valores_genero = [varones, mujeres]
etiquetas_genero = ['Varones', 'Mujeres']
colores_genero = ['skyblue', 'pink']

plt.pie(valores_genero, labels=etiquetas_genero, colors=colores_genero, autopct='%1.1f%%')
plt.title("Distribucin por Gnero")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
X, Y = np.meshgrid(np.linspace(-5, 10, 100),
                   np.linspace(-5, 10, 100))
Z = np.sqrt(X ** 2 + Y ** 2)

# Contour
fig, ax = plt.subplots()
ax.contour(X, Y, Z)

plt.show()
```
